```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
from sklearn import preprocessing
df=pd.read_csv("users.csv")
print(df.columns)
X,Y= df.iloc[:, 0:-1],df["ismalicious"]
enc = preprocessing.OrdinalEncoder()
X=enc.fit_transform(X)
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.30)
SVC_model = svm.SVC()
SVC_model.fit(X_train, Y_train)
Y_pred=SVC_model.predict(X_test)
print(metrics.accuracy_score(Y_test,Y_pred))
print(metrics.confusion_matrix(Y_test,Y_pred))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["login_attempts"], df["duration"], c=df["ismalicious"], cmap="viridis", alpha=0.7)
plt.xlabel("Login Attempts")
plt.ylabel("Duration")
plt.title("Scatter Plot of Login Attempts vs Duration")
plt.colorbar(label="Is Malicious")
plt.grid(True)
plt.show()
```

```python
# Plot histogram for duration
plt.figure(figsize=(10, 6))
plt.hist(df["duration"], bins=20, color="skyblue", edgecolor="black")
plt.xlabel("Duration")
plt.ylabel("Frequency")
plt.title("Histogram of Duration")
plt.grid(True)
plt.show()

# Plot histogram for login attempts
plt.figure(figsize=(10, 6))
plt.hist(df["login_attempts"], bins=20, color="skyblue", edgecolor="black")
plt.xlabel("Login Attempts")
plt.ylabel("Frequency")
plt.title("Histogram of Login Attempts")
plt.grid(True)
plt.show()
```

```python
# Pie chart
labels = ['Not Malicious', 'Malicious']
sizes = df['ismalicious'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart of Malicious vs Not Malicious")
plt.axis('equal')
plt.show()
```

```python
# Bar chart
plt.figure(figsize=(10, 6))
df['ismalicious'].value_counts().plot(kind='bar', color=['blue', 'red'])
plt.xlabel('Is Malicious')
plt.ylabel('Count')
plt.title('Bar Chart of Malicious vs Not Malicious')
plt.xticks([0, 1], ['Not Malicious', 'Malicious'], rotation=0)
plt.grid(axis='y')
plt.show()
```
